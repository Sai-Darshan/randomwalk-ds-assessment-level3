*__1.How many books do not have an original title [books.csv]?__*



```python
books_without_title = books["original_title"].isnull().sum()
print("No of books without an original title:", books_without_title)
```

*__2. Number of unique books after removing null values:__*


```python
books = books.dropna(subset=["original_title"])

unique_books = books["book_id"].nunique()
print("No of unique books:", unique_books)
```

*__3. Number of unique users:__*
 

```python
unique_users = ratings["user_id"].nunique()
print(f'Unique users:{unique_users}')
```

*__4. Number of unique tags:__*


```python
unique_tags = book_tags["tag_id"].nunique()
print("Unique tags:", unique_tags)
```

*__5. Most frequentlu used tag_id__*

```python
most_freq_tag_id = book_tags["tag_id"].value_counts().index[0]
print("Most frequently used tag_id:", most_freq_tag_id)
```

*__6. Book with most tag counts__*

```python
merged_data = books.merge(book_tags, on="goodreads_book_id")

most_tagged_book = merged_data.groupby("original_title")["count"].sum().sort_values(ascending=False).index[0]
print("Book with most tag counts:", most_tagged_book)
```

*__7. Plot a bar chart with top 20 unique tags in descending order of â€˜user recordsâ€™ (the number of users tagged the given tag_id with the goodreads_book_id)__*

```python
import matplotlib.pyplot as plt
```

```python
top_20_tags = book_tags.groupby("tag_id")["count"].sum().sort_values(ascending=False).head(20)
top_20_tags.plot(kind="bar")

plt.xlabel("Tag ID")
plt.ylabel("Number of User Records")

plt.title("Top 20 Tags by User Records")
plt.tight_layout()
plt.show()
```
